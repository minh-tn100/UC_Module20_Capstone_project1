# Proprocessing 

```python
import pandas as pd
import numpy as np
```

## Read CIF data and filter for 2022 data

```python
df_cif = pd.read_csv('CUST_METRICS_DEMOGRAPHIC.csv')
df_cif['CIF_OPEN_DATE'] = pd.to_datetime(df_cif['CIF_OPEN_DATE'])
df_2022 = df_cif[df_cif['CIF_OPEN_DATE'].dt.year == 2022]
del df_cif
```

## Read active customer data for 2023-10-31

```python
df_Y = pd.read_csv('CUST_METRICS_ACTIVE.csv')
df_Y = df_Y[df_Y['SYM_RUN_DATE']=='2023-10-31']
df_Y.head()
```

## Assign Y values based on LAST_ACTIVE_DATE_1M

```python
df_Y = df_Y.assign(Y = np.where(df_Y['LAST_ACTIVE_DATE_1M'].isnull(),1,0))
df_Y.head()
```

## Filter relevant columns for Y

```python
df_Y_filter = df_Y[['CLIENT_NO','Y']]
df_Y_filter
```

## Display counts and percentages

```python
# Display counts and percentages
value_counts = df_Y_filter['Y'].value_counts()
percentages = df_Y_filter['Y'].value_counts(normalize=True) * 100

result_df = pd.DataFrame({'Count': value_counts, 'Percentage': percentages})
print(result_df)
```

## Read active customer active data for 2023-09-30

```python
df_active = pd.read_csv('CUST_METRICS_ACTIVE.csv',index_col=[0])
df_active = df_active[df_active['SYM_RUN_DATE']=='2023-09-30']
df_active.drop(columns=['SYM_RUN_DATE','LAST_ACTIVE_DATE_1M'],inplace=True)
df_active.head()
```

```python
df_active.shape
```

## Read card data for 2023-09-30

```python
df_card = pd.read_csv('CUST_METRICS_CARD.csv', index_col=[0])
df_card = df_card[df_card['SYM_RUN_DATE'] == '2023-09-30']
df_card.drop(columns=['SYM_RUN_DATE'], inplace=True)
df_card.head()
```

## Read deposit spot data for 2023-09-30

```python
df_spot = pd.read_csv('CUST_METRICS_DEPO_SPOT.csv', index_col=[0])
df_spot = df_spot[df_spot['SYM_RUN_DATE'] == '2023-09-30']
df_spot.drop(columns=['SYM_RUN_DATE'], inplace=True)
df_spot.head()
```

## Read deposit transaction data for 2023-09-30

```python
df_tran = pd.read_csv('CUST_METRICS_DEPO_TRANS.csv')
df_tran.head()
```

```python
df_tran = df_tran[df_tran['SYM_RUN_DATE'] == '30/9/2023']
df_tran.drop(columns=['SYM_RUN_DATE'], inplace=True)
df_tran.head()
```

## Read e-banking transaction data for 2023-09-30

```python
df_ebank = pd.read_csv('CUST_METRICS_EBANK_TRANS.csv', index_col=[0])
df_ebank = df_ebank[df_ebank['SYM_RUN_DATE'] == '2023-09-30']
df_ebank.drop(columns=['SYM_RUN_DATE'], inplace=True)
df_ebank.head()
```

## Read lending balance data for 2023-09-30

```python
df_lend_balance = pd.read_csv('CUST_METRICS_LENDING_BALANCE_2023.csv', index_col=[0])
df_lend_balance = df_lend_balance[df_lend_balance['SYM_RUN_DATE'] == '2023-09-30']
df_lend_balance.drop(columns=['SYM_RUN_DATE'], inplace=True)
df_lend_balance.head()
```

## Read lending balance data for 2023-09-30

```python
df_balance_spot = pd.read_csv('CUST_METRICS_LENDING_SPOT.csv', index_col=[0])
df_balance_spot = df_balance_spot[df_balance_spot['SYM_RUN_DATE'] == '2023-09-30']
df_balance_spot.drop(columns=['SYM_RUN_DATE'], inplace=True)
df_balance_spot.head()
```

## Read usage data for 2023-09-30

```python
df_usage = pd.read_csv('CUST_METRICS_USAGE.csv', index_col=[0])
df_usage = df_usage[df_usage['SYM_RUN_DATE'] == '2023-09-30']
df_usage.drop(columns=['SYM_RUN_DATE'], inplace=True)
df_usage.head()
```

## Merge all dataframes

```python
full_df = pd.merge(df_2022,df_active,how='left',on='CLIENT_NO')
full_df
```

```python
full_df = full_df.merge(df_card,'left','CLIENT_NO')
full_df
```

```python
full_df = full_df.merge(df_spot,'left','CLIENT_NO')
full_df
```

```python
full_df = full_df.merge(df_tran,'left','CLIENT_NO')
full_df
```

```python
full_df = full_df.merge(df_ebank,'left','CLIENT_NO')
full_df
```

```python
full_df = full_df.merge(df_lend_balance,'left','CLIENT_NO')
full_df
```

```python
full_df = full_df.merge(df_balance_spot,'left','CLIENT_NO')
full_df
```

```python
full_df = full_df.merge(df_usage,'left','CLIENT_NO')
full_df
```

```python
full_df = full_df.merge(df_Y_filter,'left','CLIENT_NO')
full_df
```

## Filter customers aged 18 and above with non-null e-banking and mobile banking first open dates

```python
df_age_18 = full_df[full_df['AGE']>=18]
df_age_18
```

```python
df_age_18 = df_age_18[df_age_18['EBANK_FIRST_OPEN_CHANNEL'].notna()]
df_age_18 = df_age_18[df_age_18['MB2_FIRST_OPEN_DATE'].notna()]
```

## Save the filtered dataframe as a Parquet file

```python
df_age_18.to_parquet('df_age_18.parquet')
```
